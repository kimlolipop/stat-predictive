```python
from tqdm import tqdm

def decorater_process_stats(section_name:str, test_df: pd.DataFrame) -> pd.DataFrame:
  INDEX_TEMP_TO_CURRENT = (2,15)
  columns = test_df.keys()[INDEX_TEMP_TO_CURRENT[0]:INDEX_TEMP_TO_CURRENT[1]]
  status_json_list = []
  for row in tqdm(test_df.iterrows()):
    for column in columns:
      sensor_split = column.split('_')
      if column == 'temperature':
        sensorname,sensornumber = 'temperature','1'
      else:
        sensorname, sensornumber = sensor_split[0],sensor_split[1]

      sensorvalue = row[1][column]
      status = process_stats_model(section_name, sensorname, sensornumber, sensorvalue)
      status_json = {'date':row[1]['date'], 'time':row[1]['time'], 'sectionname':section_name, 'ai_status':status, 'sensor': column}
      status_json_list.append(status_json)
      status_df = pd.DataFrame(data=status_json_list)
  return status_df

result_df = decorater_process_stats('R1L1-J3',test_df)
result_df.to_csv('J3_tranform.csv')
result_df
```
